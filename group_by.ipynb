# Group By


groupby, bir DataFrame’i belirli bir sütuna göre gruplandırarak her grup için özet istatistikler 
(toplam, ortalama, sayım vb.) üretmeye yarayan bir Pandas fonksiyonudur.

groupby() temel olarak üç aşamada çalışan bir süreçtir:

*Bölme (Splitting): Veri kümesi, bir veya daha fazla sütundaki benzersiz değerlere göre ayrı gruplara ayrılır.

*Uygulama (Applying): Her bir alt gruba (split edilen parçaya) bağımsız olarak bir hesaplama (toplama, ortalama, sayma, standart sapma vb. gibi bir toplama fonksiyonu) uygulanır.

*Birleştirme (Combining): Tüm gruplara uygulanan hesaplamaların sonuçları, nihai bir sonuç nesnesi (genellikle yeni bir DataFrame) halinde birleştirilir.

groupby grupları oluşturur.
agg ise bu gruplar üzerinde ne işlem yapılacağını söyler.


toplam almak → "sum"

ortalama → "mean"

en büyük değer → "max"

en küçük → "min"

sayma → "count"

```python
import pandas as  pd
```

```python
df = pd.read_csv(
    filepath_or_buffer='Data/imdb.csv'
)
```

```python
df.columns
```

```python
# Yıllara göre toplam Num_Reviews ler nelerdir?

(
    df.groupby(by='YR_Released')
    .agg(
        {
            'Num_Reviews' : 'sum',
        }
    )
    .query('Num_Reviews >8000000 & Num_Reviews <15000000')
    .sort_values(by='Num_Reviews' , ascending=False)
)
```

```python
# Yıllara göre toplamda alınan rating ler nelerdir?

(
    df.groupby(by='YR_Released')
    .agg(
        {
            'Rating' : 'sum'
        }
    )
    .sort_values(by='Rating' , ascending=False)
)
```

```python
# Ratinglere göre Num_Reviews görüntülenmelerin toplamını gruplayan bir sorgu yazınız.

(
    df.groupby(by='Rating')
    .agg(
        {
            'Num_Reviews' : 'sum'
        }
    )
    .sort_values(by='Num_Reviews' , ascending=False)
)
```

```python
# Hangi yıl en yüksek ortalama rating vardır?


# (
#     df.groupby(by='YR_Released')['Rating']
#     .mean()
#     .sort_values(ascending=False)
# ).head(10)


# .agg ile yazmak istersek:

(
    df.groupby(by='YR_Released')
    .agg(
        avg_rating=('Rating', "mean")
        )
        .sort_values(by="avg_rating", ascending=False)
)
```

```python
# En çok film hangi yıl  çekilmiş?

(
    df.groupby(by='YR_Released')
    ['Movie_Title']
    .count()
    .sort_values(ascending=False)
    .head(10)
)
```
