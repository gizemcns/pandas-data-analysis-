# DATA FILTERING


# Data Filtering (Veri Filtreleme) Nedir?
Tanım:
Data filtering, veriyi koşullara göre süzerek yalnızca analiz için gerekli kısmını elde etme işlemidir.

İşlev:
Veri filtrelemenin temel işlevi, veri setini analiz amacına uygun hale getirmektir.

Amaç:
Veriyi analiz öncesi sadeleştirmek, odaklanmak istenen bilgiye ulaşmak ve performansı artırmaktır.

Yöntem:
Pandas'ta filtreleme için birden fazla yöntem kullanılır, ancak en yaygın olanları şunlardır:

Mantıksal Koşullar (Boolean Indexing):

Bir veya daha fazla sütuna uygulanan karşılaştırma operatörleri (>, <, ==, !=, vb.) ve mantıksal operatörler (& (VE), | (VEYA), ~ (DEĞİL)) kullanılarak yapılanlar.

Örnek: df[df['Yaş'] > 30]

.loc ve .iloc Indexleyiciler:

.loc (etiket tabanlı): Satır ve sütunları etiket adlarıyla seçerken mantıksal koşullar da uygulayabilir.

.iloc (konum tabanlı): Satır ve sütunları tam sayı konumlarıyla seçer (filtreleme için daha az kullanılır, ancak alt kümeleme için önemlidir).

.isin() Metodu:

Bir sütundaki değerlerin, bir listede bulunan değerlerden herhangi biri olup olmadığını kontrol eder. Çoklu değer kontrolü için çok etkilidir.

Örnek: df[df['Şehir'].isin(['Ankara', 'İstanbul'])]

.str Erişimcisi:

Dize (string) tabanlı sütunlarda metin filtrelemesi yapmak için kullanılır (örneğin, belirli bir kelimeyi içeren veya belirli bir harfle başlayan satırlar).

Örnek: df[df['Ürün Adı'].str.contains('Premium')]

```python
# todo: YR_Released sütunundan 2000 yılından sonra yayınlanmış filmlerin Movie_Title YR_Released , Rating sütunlarındaki verileri çeken bir sorgu yazın.
# todo: ayrıca Rating satırında bulunan verilere göre puanı en yüksekten sen düşüğe olacak şekilde sıralayın. select yaptık sutun seçtik sonrada sıraladık. bu veriseti için rapourumsu bir sorgu yazdık
(
    df.query('YR_Released > 2000')
        .loc[: , ['Movie_Title' , 'YR_Released' , 'Rating']]
        .sort_values(by='Rating', ascending= False)
)

# genelde filtre attığın değere göre sıralarsın. Yani burada sort atarken YR_Released e göre atılmalı biz yapmadık. 
```

```python
import time

start = time.time_ns()

(
    df.query('YR_Released > 2000')
        .loc[: , ['Movie_Title' , 'YR_Released' , 'Rating']]
        .sort_values(by='Rating', ascending= False)
)

finish = time.time_ns()

print(finish - start )
```

```python
# her iki kodda aynı işi yapar query daha okunaklıdır.Bu daha karısıktır.

df[df['YR_Released'] > 2000][['Movie_Title', 'YR_Released' , 'Rating']].sort_values(by='Rating', ascending= False)

```

```python
import time

start1 = time.time_ns()

df[df['YR_Released'] > 2000][['Movie_Title', 'YR_Released' , 'Rating']].sort_values(by='Rating', ascending= False)


finish1 = time.time_ns()

print(finish1 - start1 )
```

method chaining (veya zincirleme metot kullanımı)  query-style filtering daha teknik söylemiyle expression-based filtering ile yazmaktan daha kısa sürmektedir.

```python
# ? Rating sütunundaki verilerin 7.0 ile 9.0 arasındaki verileri içeren 'Movie_Title','YR_Released', 'Rating' sütunlarını listeleyin.

(
    df.query('Rating > 7.0 and Rating < 9.0')
    .loc[: , ['Movie_Title' , 'YR_Released' , 'Rating']]
        .sort_values(by='Rating', ascending= False)
)

# and yazarken hem & hem de and yazabiliriz.
```

```python
# ? Num_Reviews >30000 den büyük ve YR_Released >2000 olan filmlerin Movie_Title, Num_Reviews, YR_Released sütunlarını, Num_Reviews  
# ? e göre çoktan aza sıralancak şekilde listeleyin. 
(
    df.query('Num_Reviews > 30000 and YR_Released >2000')
    .loc[: , ['Movie_Title' ,'Num_Reviews' , 'YR_Released' ]]
        .sort_values(by='Num_Reviews', ascending= False)
)
```

```python
# ? Num_Reviews 1808543 ya da 1599451 ya da 1332276 olan filmlerin Movie_Title, Num_Reviews, Rating sütunlarını, Num_Reviews  
# ? e göre çoktan aza sıralancak şekilde listeleyin.

(
    df.query('Num_Reviews == 1808543 | Num_Reviews == 1599451 | Num_Reviews == 1332276')
    .loc[: , ['Movie_Title' ,'Num_Reviews' , 'Rating' ]]
        .sort_values(by='Num_Reviews', ascending= False)
)

# & → AND

# | → OR

# ~ → NOT
```

```python
# ? Movie_Title içerisinde 'King' ifadesi geçen ve YR_Released >2000 olan filmlerin, Movie_Title, Num_Reviews, YR_Released sütunlarını, Num_Reviews  
# ? e göre  azdan çoka sıralancak şekilde listeleyin. 
```

```python
(
df.query('Movie_Title.str.contains("King") & YR_Released >2000 ') 
    .loc[:, ['Movie_Title', 'Num_Reviews' , 'YR_Released']] 
    .sort_values(by='Num_Reviews', ascending=True)

)
```

```python
df[(df['Movie_Title'].str.contains("King")) & (df['YR_Released'] > 2000)][['Movie_Title', 'Num_Reviews', 'YR_Released']].sort_values(by='Num_Reviews', ascending=True) 
```

# .loc & .iloc

.loc vs .iloc nedir? 

Pandas'taki .loc[] ve .iloc[] özellikleri, bir Pandas DataFrame'indeki belirli satırlara ve sütunlara erişmek (veya bir veri kümesini dilimlemek) için kullanılır.

.loc komutu etiket tabanlıdır. 

.iloc komutu ise integer tabanlıdır. Yani .loc komutunu kullanırken satır ya da sütun adı söyleriz fakat .iloc komutunda o satır veya sütunun numarasını belirtiriz. iloc index bazlı olarak ilerler.

.loc[] özelliği etiket indekslemesi için kullanılırken, .iloc[] özelliği tam sayı indekslemesi için kullanılır.
